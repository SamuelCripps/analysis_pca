```python
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.datasets import load_breast_cancer
```

```python
cancer = load_breast_cancer()
```

```python
cancer.keys()
```

```python
print(cancer['DESCR'])
```

```python
 df = pd.DataFrame(cancer['data'], columns = cancer['feature_names'])
```

```python
df.head(5)
```

from the above table, since the difference between the values is very high like in 1000s, we need to normalize them and hance use the standard scaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
scaler.fit(df)
```

```python
#this step is required to finish the scaling
scaled_data = scaler.transform(df)
```

```python
scaled_data
```

```python
from sklearn.decomposition import PCA
```

**PCA is basically used for dimension reduction.**

```python
pca = PCA(n_components=2)
```

```python
pca.fit(scaled_data)
#because of more components there might be underfitting problem for the model. 
```

```python
x_pca = pca.transform(scaled_data)
```

```python
scaled_data.shape
```

```python
x_pca.shape
```

```python
scaled_data
```

```python
x_pca
```

```python
plt.figure(figsize=(10,8))
plt.scatter(x_pca[:,0],x_pca[:,1],c=cancer['target'])
plt.xlabel('First principle component')
plt.ylabel('Second principle component')
```
